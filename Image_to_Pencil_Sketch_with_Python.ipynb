**IMPORTING ALL THE REQUIRED LIBRARIES**

```python
import cv2
import io # io Library used for various input and ouput operations
from google.colab import files # this library used for dealing with files
from PIL import Image
from IPython.display import display
import matplotlib.pyplot as plt
%matplotlib inline
```

**DISPLAYING THE ORIGINAL IMAGE**

```python
img=cv2.imread('/content/dhoni.jpg')

img = cv2.resize(img, (600, 500))
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

**DISPLAYING THE BACK AND WHITE IMAGE**

```python
plt.imshow(img, cmap='gray')
```

```python
img.max()
```

```python
img.min()
```

```python
inv = 255-img
```

```python
plt.imshow(inv, cmap='gray')
```

```python
blur = cv2.GaussianBlur(inv, ksize=(21,21), sigmaX=0, sigmaY=0)
```

```python
plt.imshow(blur, cmap='gray')
```

```python
dodge = lambda image, mask: cv2.divide(image, 255-mask, scale=256)
```

```python
blended = dodge(img, blur)
```

**DISPLAYING THE SKECTH IMAGE**

```python
plt.imshow(blended, cmap='gray')
```

```python
cv2.imwrite('imgsketch.jpg', blended)
```
